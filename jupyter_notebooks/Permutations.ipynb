```python
'''Design an algorithm to print out all the permuations of a string'''
def permutations(s):
    per=[]
    if len(s) == 1:
        per.append(s)
    else:
        for i in range(len(s)):
            letter = s[i]
            remainingList = s[:i] + s[i+1:]
            ##print type(letter)
            for p in permutations(remainingList):
                ##print type(p)
                per.append(letter + p)
    return per
```

```python
s = 'abcdeh'
permu = permutations(s)
```
